```python
import whisper

whisper_model = whisper.load_model("base")
```

```python
data = torchaudio.datasets.LIBRITTS(root="data", url="test-clean", download=True)
# data_loader = torch.utils.data.DataLoader(data, batch_size=1, shuffle=True)
```

```python
import editdistance

char_edit_dist = []
word_edit_dist = []

for i in range(len(data)):
  item = data[i]

  a, b, c, d = item[6].split("_")
  file_path = f"data/LibriTTS/test-clean/{a}/{b}/{a}_{b}_{c}_{d}.wav"

  output = whisper_model.transcribe(file_path)["text"]
  label = item[3]

  char_edit_dist.append(editdistance.eval(output, label))
  word_edit_dist.append(editdistance.eval(output.split(" "), label.split(" ")))

  if i % 10 == 9:
    print()
    print(f"After {i+1} testing examples...")

    average_char_edit_dist = sum(char_edit_dist)/len(char_edit_dist)
    print("Average char edit dist: ", average_char_edit_dist)

    average_word_edit_dist = sum(word_edit_dist)/len(word_edit_dist)
    print("Average word edit dist: ", average_word_edit_dist)
```
